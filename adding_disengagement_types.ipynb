```python
from collections import Counter
all_disengagement_types = [item for sublist in disengagement_reasons_list for item in sublist]

# Count occurrences of each type
disengagement_counts = Counter(all_disengagement_types)

# Print the counts
print("Disengagement type counts:")
for disengagement_type, count in disengagement_counts.items():
    print(f"{disengagement_type}: {count}")
```

```python
# Map disengagement reasons to integers
reason_mapping = {
    'OBS': 0,
    'Safety': 1,
    'Turnback': 2,
    'Pedestrian crossing': 3,
    'Localization': 4,
    'Bad engage': 5,
    'Give way': 6,
    'SPEED': 7,
    'STOP': 8,
}
```

```python
base_dir = "F:\\"

# Iterate through files and process them
for file, disengagement_types, disengagement_reasons in zip(unified_files, disengagement_types_list, disengagement_reasons_list):
    file_path = f"{base_dir}{file}"
    df = pd.read_csv(file_path)
    
    # Ensure drivemode column is numeric
    df['drivemode'] = pd.to_numeric(df['drivemode'], errors='coerce')
    
    # Identify disengagements
    df['disengagement'] = (df['drivemode'].shift(1) == 1.0) & (df['drivemode'] == 0.0)
    
    # Initialize disengagement_type column
    df['disengagement_type'] = None
    df['disengagement_reason'] = None
    
    # Assign planned/unplanned types to disengagements
    disengagement_indices = df.index[df['disengagement']].tolist()
    for i, disengagement_index in enumerate(disengagement_indices):
        if i < len(disengagement_types):  # Use the provided list
            df.at[disengagement_index, 'disengagement_type'] = disengagement_types[i]
            df.at[disengagement_index, 'disengagement_reason'] = disengagement_reasons[i]
        else:  # Default to unplanned (0) if list is shorter
            df.at[disengagement_index, 'disengagement_type'] = 0
            df.at[disengagement_index, 'disengagement_reason'] = 'OBS'

    # For timestamps that do not have a disengagement, fill in the missing values with -1
    df['disengagement_reason'] = df['disengagement_reason'].map(reason_mapping).fillna(-1).astype(int)
    
    # Generate output filename
    output_filename = file.replace('_unified_timestamps.csv', '_with_disengagements.csv')
    output_path = f"{base_dir}{output_filename}"
    
    # Saving
    df.to_csv(output_path, index=False)

    print(f"Processed and saved: {output_path}")
```
